```
import cv2
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
```

```
temp_image = cv2.imread('/content/drive/MyDrive/Datasets/Output_frames/85.png')
```

```
temp_image.shape
```

```
#calling the model
bbox,label, conf = cv.detect_common_objects(temp_image, model = 'yolov3')
```

```
output_image = draw_bbox(temp_image, bbox, label, conf)
```

```
plt.figure(figsize=(16,9))
plt.imshow(output_image)
plt.show()
print('Number of cars in the image is ' + str(label.count('car')))
```
